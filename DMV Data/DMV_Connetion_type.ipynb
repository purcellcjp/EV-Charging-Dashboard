```python
#import commands
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Import CSV files
DMV_Merge = Path("DMV_Merge.csv")
```

```python
# Read the files and store the data
DMV_Merge_data = pd.read_csv(DMV_Merge)
```

```python
#Select only Station_Name, ID, and Connection Type fields
DMV_Fields = DMV_Merge_data[['ID', 'Connection_Type']]
DMV_Fields
```

```python
# Group By ID and Connection_type
DMV_Fields = DMV_Merge_data[['ID', 'Connection_Type']].copy()
Connect_list = DMV_Fields.groupby('ID')['Connection_Type'].apply(', '.join).reset_index()
print(Connect_list)
```

```python
# Step 1: Group by ID and create a comma-delimited string
Connect_list = DMV_Merge_data.groupby('ID')['Connection_Type'].apply(', '.join).reset_index()

# Step 2: Define the tidy_split function
def tidy_split(df, column, sep=' ', keep=False):
    indexes = []
    new_values = []
    df = df.dropna(subset=[column])
    for i, presplit in enumerate(df[column].astype(str)):
        # Split on both comma and space to handle different formats
        values = [v.strip() for v in presplit.replace(',', ' ').split()]
        if keep and len(values) > 1:
            indexes.append(i)
            new_values.append(presplit)
        for value in values:
            indexes.append(i)
            new_values.append(value)
    new_df = df.iloc[indexes, :].copy()
    new_df[column] = new_values
    return new_df

# Step 3: Use tidy_split on the Connect_list DataFrame
tidy_df = tidy_split(Connect_list, 'Connection_Type')

# Step 4: Rename columns
tidy_df.rename(columns={'ID': 'Fuel_Station_Id', 'Connection_Type': 'Connector_Type_Code'}, inplace=True)

# Display the result
print(tidy_df)
```

```python
#import commands
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
#import csv files
alt_fuel_load = Path("Data/alt_fuel_stationsDC.csv")
alt_fuel_load2 = Path("Data/alt_fuel_stationsMD.csv")
alt_fuel_load3 = Path("Data/alt_fuel_stationsVA.csv")
```

```python
# Read Data File and store into Pandas DataFrames
DC_fuel_data = pd.read_csv(alt_fuel_load)
MD_fuel_data = pd.read_csv(alt_fuel_load2)
Va_fuel_data = pd.read_csv(alt_fuel_load3)
```

```python
# Grab selected fields
DC_Fields = DC_fuel_data[['Fuel Type Code','Station Name', 'Street Address', 'Date Last Confirmed','Groups With Access Code', 'State', 'City', 'Open Date', 'ZIP', 'EV Connector Types', 'EV Pricing', 'Latitude', 'Longitude','Access Code']]
DC_Fields.head()
```

```python
# add ID field
DC_Fields.loc[:, 'ID'] = range(1, len(DC_Fields) + 1)
DC_Fields.head()
```

```python
# Grab selected fields
MD_Fields = MD_fuel_data[['Fuel Type Code','Station Name', 'Street Address','Date Last Confirmed','Groups With Access Code','State', 'City', 'Open Date', 'ZIP', 'EV Connector Types', 'EV Pricing', 'Latitude', 'Longitude','Access Code']]
MD_Fields.head()
```

```python
# Add ID field
MD_Fields.loc[:, 'ID'] = range(1, len(MD_Fields) + 1)
MD_Fields.head()
```

```python
# Grab Selected Fields
Va_fields = Va_fuel_data[['Fuel Type Code','Station Name', 'Street Address','Date Last Confirmed','Groups With Access Code', 'State', 'City', 'Open Date', 'ZIP', 'EV Connector Types', 'EV Pricing', 'Latitude', 'Longitude','Access Code']]
Va_fields.head()
```

```python
Va_fields.loc[:, 'ID'] = range(1, len(Va_fields) + 1)
Va_fields.head()
```

```python
# clean Data
DC_Fuel_cleaned = DC_Fields.dropna()
DC_Fuel_cleaned.head()
```

```python
# clean Data
MD_Fuel_Cleaned = MD_Fields.dropna()
MD_Fuel_Cleaned.head()
```

```python
Va_Fuel_Cleaned = Va_fields.dropna()
Va_Fuel_Cleaned.head()
```

```python
# combine data tables
DMV_Fuel_Data = pd.concat([DC_Fuel_cleaned, MD_Fuel_Cleaned, Va_Fuel_Cleaned],  ignore_index=True)
DMV_Fuel_Data
```

```python
# fixing State field 
DMV_Fields = DMV_Fuel_Data [['State']]
unique_states = DMV_Fields['State'].unique()
print(unique_states)
```

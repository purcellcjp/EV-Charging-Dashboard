```python
#import commands
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
#import csv files
Data_csv = Path("Merged Data/data.csv")
DMV_Fuel_Data = Path("Merged Data/DMV_Fuel_Data.csv")
```

```python
# Read Data File and store into Pandas DataFrames
Dawson_data = pd.read_csv(Data_csv)
Daniel_data = pd.read_csv(DMV_Fuel_Data)
```

```python
# print Dawson Data
Dawson_data.head()
```

```python
# Select fields
Fields = Dawson_data[['ID','Station Name','Access','Owner','Last_Used', 'Open_Date','Address', 'City','State','Zipcode','Latitude','Longitude','Pricing', 'Connection_Type' ]]
Fields.head()
```

```python
# Daniels Data
Daniel_data.head()
```

```python
# fields
Fields_2 = Daniel_data[['ID', 'Fuel Type Code','Station Name', 'Street Address', 'State', 'City', 'Open Date', 'Date Last Confirmed', 'ZIP', 'EV Connector Types', 'EV Pricing', 'Latitude', 'Longitude','Access Code']]
Fields_2.head()
```

```python
# Rename columns
Fields_2.rename(columns={
    'Street Address': 'Address',
    'Open Date': 'Open_Date',
    'Date Last Confirmed': 'Last_Used',
    'ZIP': 'Zipcode',
    'EV Connector Types': 'Connection_Type',
    'EV Pricing': 'Pricing',  
    'Access Code': 'Access'
}, inplace=True)
Fields_2.head()
```

```python
Daniel_data.rename(columns={
    'Street Address': 'Address',
    'Open Date': 'Open_Date',
    'Date Last Confirmed': 'Last_Used',
    'ZIP': 'Zipcode',
    'EV Connector Types': 'Connection_Type',
    'EV Pricing': 'Pricing',  
    'Access Code': 'Access'
}, inplace=True)
Daniel_data.head()

```

```python
#merge data frames
DMV_Merge = pd.merge(Dawson_data, Daniel_data, on='Access', how='inner')
DMV_Merge.head()
```

```python
# create output path csv file
output_path = Path("DMV_Merge.csv")
DMV_Merge.to_csv(output_path, index=False)
```

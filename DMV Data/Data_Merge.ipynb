```python
#import commands
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
#import csv files
Data_csv = Path("Merged Data/data.csv")
DMV_Fuel_Data = Path("Merged Data/DMV_Fuel_Data.csv")
```

```python
# Read Data File and store into Pandas DataFrames
Dawson_data = pd.read_csv(Data_csv)
Daniel_data = pd.read_csv(DMV_Fuel_Data)
```

```python
# print Dawson Data
Dawson_data.head()
```

```python
# Select fields
Fields = Dawson_data[['ID','Station Name','Access','Owner','Last_Used', 'Open_Date','Address', 'City','State','Zipcode','Latitude','Longitude','Pricing', 'Connection_Type' ]]
Fields.head()
```

```python
# Daniels Data
Daniel_data.head()
```

```python
# fields
Fields_2 = Daniel_data[['Fuel Type Code','Station Name', 'Street Address','Date Last Confirmed','Groups With Access Code', 'State', 'City', 'Open Date', 'ZIP', 'EV Connector Types', 'EV Pricing', 'Latitude', 'Longitude','Access Code', 'Facility Type', 'EV Level1 EVSE Num',	'EV Level2 EVSE Num', 'EV DC Fast Count'
]]
Fields_2.head()
```

```python
# Rename columns
Fields_2.rename(columns={
    'Street Address': 'Address',
    'Open Date': 'Open_Date',
    'Date Last Confirmed': 'Last_Used',
    'ZIP': 'Zipcode',
    'EV Connector Types': 'Connection_Type',
    'EV Pricing': 'Pricing',  
    'Access Code': 'Access',  
    'Facility Type': 'Owner',
    'EV Level1 EVSE Num': 'Level1_Charging',
    'EV Level2 EVSE Num': 'Level2_Charging', 
    'EV DC Fast Count': 'DC_Fast'
}, inplace=True)
Fields_2.head()
```

```python
Daniel_data.rename(columns={
    'Street Address': 'Address',
    'Open Date': 'Open_Date',
    'Date Last Confirmed': 'Last_Used',
    'ZIP': 'Zipcode',
    'EV Connector Types': 'Connection_Type',
    'EV Pricing': 'Pricing',  
    'Access Code': 'Access',  
    'Facility Type': 'Owner',
    'EV Level1 EVSE Num': 'Level1_Charging',
    'EV Level2 EVSE Num': 'Level2_Charging', 
    'EV DC Fast Count': 'DC_Fast'
}, inplace=True)
Daniel_data.head()

```

```python
# data type for both data sets
Dawson_data.dtypes
Daniel_data.dtypes
```

```python
# fix the different data set types for ID
Dawson_data['ID'] = Dawson_data['ID'].astype(str)
Daniel_data['ID'] = Daniel_data['ID'].astype(str)
```

```python
#merge data frames
merge_columns = [
    'ID', 'Station Name', 'Access', 'Owner', 'Last_Used', 'Open_Date',
    'DC_Fast', 'Level1_Charging', 'Level2_Charging', 'Address', 'City',
    'State', 'Zipcode', 'Latitude', 'Longitude', 'Pricing',
    'Connection_Type'
]

if 'Owner' not in Daniel_data.columns:
    Daniel_data['Owner'] = None 

DMV_Merge = pd.merge(Dawson_data, Daniel_data, on=merge_columns, how='outer')
DMV_Merge.head()
```

```python
# check for duplicates 
duplicate_count = DMV_Merge.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

```python
# create output path csv file
output_path = Path("DMV_Merge.csv")
DMV_Merge.to_csv(output_path, index=False)
```

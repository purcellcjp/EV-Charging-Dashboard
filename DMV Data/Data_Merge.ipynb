```python
#import commands
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
#import csv files
Data_csv = Path("Merged Data/data.csv")
DMV_Fuel_Data = Path("Merged Data/DMV_Fuel_Data.csv")
```

```python
# Read Data File and store into Pandas DataFrames
Dawson_data = pd.read_csv(Data_csv)
Daniel_data = pd.read_csv(DMV_Fuel_Data)
```

```python
# fields
Fields_2 = Daniel_data[['Fuel Type Code','Station Name', 'Street Address','Date Last Confirmed','Groups With Access Code', 'State', 'City', 'Open Date', 'ZIP', 'EV Connector Types', 'EV Pricing', 'Latitude', 'Longitude','Access Code', 'Facility Type', 'EV Level1 EVSE Num',	'EV Level2 EVSE Num', 'EV DC Fast Count'
]]
Fields_2.head()
```

```python
# Rename columns in Fields_2 to standard names
Fields_2.rename(columns={
    'Street Address': 'Address',
    'Open Date': 'Open_Date',
    'Date Last Confirmed': 'Last_Used',
    'ZIP': 'Zipcode',
    'EV Connector Types': 'Connection_Type',
    'EV Pricing': 'Pricing',  
    'Access Code': 'Access',  
    'Facility Type': 'Owner',
    'EV Level1 EVSE Num': 'Level1_Charging',
    'EV Level2 EVSE Num': 'Level2_Charging', 
    'EV DC Fast Count': 'DC_Fast'
}, inplace=True)
# Display the first few rows of the updated Fields_2 DataFrame
Fields_2.head()
```

```python
# Rename columns in Daniel_data to standardize names
Daniel_data.rename(columns={
    'Street Address': 'Address',                  # Rename 'Street Address' to 'Address'
    'Open Date': 'Open_Date',                     # Rename 'Open Date' to 'Open_Date'
    'Date Last Confirmed': 'Last_Used',           # Rename 'Date Last Confirmed' to 'Last_Used'
    'ZIP': 'Zipcode',                             # Rename 'ZIP' to 'Zipcode'
    'EV Connector Types': 'Connection_Type',      # Rename 'EV Connector Types' to 'Connection_Type'
    'EV Pricing': 'Pricing',                       # Rename 'EV Pricing' to 'Pricing'
    'Access Code': 'Access',                       # Rename 'Access Code' to 'Access'
    'Facility Type': 'Owner',                      # Rename 'Facility Type' to 'Owner'
    'EV Level1 EVSE Num': 'Level1_Charging',      # Rename 'EV Level1 EVSE Num' to 'Level1_Charging'
    'EV Level2 EVSE Num': 'Level2_Charging',      # Rename 'EV Level2 EVSE Num' to 'Level2_Charging'
    'EV DC Fast Count': 'DC_Fast'                  # Rename 'EV DC Fast Count' to 'DC_Fast'
}, inplace=True)

# Display the first few rows of the updated Daniel_data DataFrame
Daniel_data.head()
```

```python
# data type for both data sets
Dawson_data.dtypes
Daniel_data.dtypes
```

```python
# fix the different data set types for ID
Dawson_data['ID'] = Dawson_data['ID'].astype(str)
Daniel_data['ID'] = Daniel_data['ID'].astype(str)
```

```python
# Replace NaN values with 0 and convert charging level columns to integers in Dawson_data
Dawson_data['Level1_Charging'] = Dawson_data['Level1_Charging'].fillna(0).astype(int)
Dawson_data['Level2_Charging'] = Dawson_data['Level2_Charging'].fillna(0).astype(int)
Dawson_data['DC_Fast'] = Dawson_data['DC_Fast'].fillna(0).astype(int)

# Replace NaN values with 0 and convert charging level columns to integers in Daniel_data
Daniel_data['Level1_Charging'] = Daniel_data['Level1_Charging'].fillna(0).astype(int)
Daniel_data['Level2_Charging'] = Daniel_data['Level2_Charging'].fillna(0).astype(int)
Daniel_data['DC_Fast'] = Daniel_data['DC_Fast'].fillna(0).astype(int)
```

```python
# Define the columns to merge on
merge_columns = [
    'ID', 'Station Name', 'Access', 'Owner', 'Last_Used', 'Open_Date',
    'DC_Fast', 'Level1_Charging', 'Level2_Charging', 'Address', 'City',
    'State', 'Zipcode', 'Latitude', 'Longitude', 'Pricing',
    'Connection_Type'
]

# Add 'Owner' column to Daniel_data if it doesn't exist
if 'Owner' not in Daniel_data.columns:
    Daniel_data['Owner'] = None 

# Merge Dawson_data and Daniel_data using specified columns and outer join
DMV_Merge = pd.merge(Dawson_data, Daniel_data, on=merge_columns, how='outer')

# Drop unnecessary columns if they exist
DMV_Merge.drop(columns=['Fuel Type Code', 'Groups With Access Code'], inplace=True, errors='ignore')

# Reset index of the merged DataFrame
DMV_Merge.reset_index(drop=True, inplace=True)

# Create a new 'ID' column with sequential numbers
DMV_Merge['ID'] = range(len(DMV_Merge))

# Display the first few rows of the merged DataFrame
DMV_Merge.head()
```

```python
# check for duplicates 
duplicate_count = DMV_Merge.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

```python
# create output path csv file
output_path = Path("DMV_Merge.csv")
DMV_Merge.to_csv(output_path, index=False)
```

```python
# Convert 'Open_Date' to datetime format
DMV_Merge['Open_Date'] = pd.to_datetime(DMV_Merge['Open_Date'])
```

```python
# Extract the year from 'Open_Date'
DMV_Merge['Year'] = DMV_Merge['Open_Date'].dt.year

# Count the number of entries for each year
year_counts = DMV_Merge['Year'].value_counts().sort_index()
```

```python
import matplotlib.pyplot as plt

# Create the line graph
plt.figure(figsize=(10, 5))  # Set figure size
plt.plot(year_counts.index, year_counts.values, marker='o')

# Add titles and labels
plt.title('Number of Stations Opened by Year')
plt.xlabel('Year')
plt.ylabel('Number of Stations')

# Show grid
plt.grid()

# Show the plot
plt.xticks(year_counts.index)  # Set x-ticks to be the years
plt.tight_layout()  # Adjust layout
plt.show()
```
